```python
import tensorflow as tf
import pandas as pd

# Load training data from CSV file
train_data = pd.read_csv('server_mon.csv')

# Preprocess timestamp column to extract hour of day
train_data['hour'] = pd.to_datetime(train_data['timestamp']).dt.hour

# Split data into input (X) and output (y) variables
X_train = train_data.drop(columns=['timestamp', 'is_anomaly'])
y_train = train_data['is_anomaly']

# Convert data to NumPy arrays and float32 data type
X_train = X_train.to_numpy().astype('float32')
y_train = y_train.to_numpy().astype('float32')

# Define neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=[tf.keras.metrics.BinaryAccuracy()])

# Train model for 100 epochs
model.fit(X_train, y_train, epochs=100)

# Load test data from CSV file
test_data = pd.read_csv('server_logs_test.csv')

# Preprocess timestamp column to extract hour of day
test_data['hour'] = pd.to_datetime(test_data['timestamp']).dt.hour

# Split data into input (X) and output (y) variables
X_test = test_data.drop(columns=['timestamp', 'is_anomaly'])
y_test = test_data['is_anomaly']

# Convert data to NumPy arrays and float32 data type
X_test = X_test.to_numpy().astype('float32')
y_test = y_test.to_numpy().astype('float32')

# Predict anomalies using trained model
y_pred = model.predict(X_test)

# Generate binary predictions based on probability threshold
y_pred_binary = (y_pred > 0.5)

# Calculate accuracy of anomaly detection
accuracy = tf.keras.metrics.BinaryAccuracy()
accuracy.update_state(y_test, y_pred_binary)
print('Accuracy: {}'.format(accuracy.result().numpy()))
```

```python
import tensorflow as tf
import pandas as pd
import numpy as np

# Load training data from CSV file
train_data = pd.read_csv('server_mon.csv')

# Preprocess timestamp column to extract hour of day
train_data['hour'] = pd.to_datetime(train_data['timestamp']).dt.hour

# Split data into input (X) and output (y) variables
X_train = train_data.drop(columns=['timestamp', 'is_anomaly'])
y_train = train_data['is_anomaly']

# Convert data to NumPy arrays and float32 data type
X_train = X_train.to_numpy().astype('float32')
y_train = y_train.to_numpy().astype('float32')

# Reshape input data for RNN
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))

# Define RNN model
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(64, activation='relu', input_shape=(1, X_train.shape[2])),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile model with binary cross-entropy loss and Adam optimizer
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=[tf.keras.metrics.BinaryAccuracy()])

# Train model for 100 epochs
model.fit(X_train, y_train, epochs=100)

# Load test data from CSV file
test_data = pd.read_csv('server_logs_test.csv')

# Preprocess timestamp column to extract hour of day
test_data['hour'] = pd.to_datetime(test_data['timestamp']).dt.hour

# Split data into input (X) and output (y) variables
X_test = test_data.drop(columns=['timestamp', 'is_anomaly'])
y_test = test_data['is_anomaly']

# Convert data to NumPy arrays and float32 data type
X_test = X_test.to_numpy().astype('float32')
y_test = y_test.to_numpy().astype('float32')

# Reshape input data for RNN
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Predict anomalies using trained model
y_pred = model.predict(X_test)

# Generate binary predictions based on probability threshold
y_pred_binary = (y_pred > 0.5)

# Calculate accuracy of anomaly detection
accuracy = tf.keras.metrics.BinaryAccuracy()
accuracy.update_state(y_test, y_pred_binary)
print('Accuracy: {}'.format(accuracy.result().numpy()))
```
